```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams

mh_df = pd.read_csv('../depression_data.csv')

mh_df.head()
```

```python
mh_df.info()
```

```python
ynmap = {'No': 0, 'Yes': 1}

mh_df['Chronic Medical Conditions'] = mh_df['Chronic Medical Conditions'].map(ynmap)
mh_df['Family History of Depression'] = mh_df['Family History of Depression'].map(ynmap)
mh_df.head()
```

```python
mh_df.info()
```

```python
mh_df['History of Substance Abuse'] = mh_df['History of Substance Abuse'].map(ynmap)
mh_df['History of Mental Illness'] = mh_df['History of Mental Illness'].map(ynmap)


mh_df
```

```python
mh_df['Sleep Patterns'].value_counts()
```

```python
sleepmap = {'Poor': 0, 'Fair': 1, 'Good': 2}

mh_df['Sleep Patterns'] = mh_df['Sleep Patterns'].map(sleepmap)

#mh_df['Sleep Patterns'] 
```

```python
mh_df.rename(columns={'Sleep Patterns': 'Sleep Quality'}, inplace=True)

mh_df
```

```python
mh_df['Dietary Habits'].value_counts()

mh_df.rename(columns={'Dietary Habits': 'Dietary Quality'}, inplace=True)

dietmap = {'Unhealthy': 0, 'Moderate': 1, 'Healthy': 2}

mh_df['Dietary Quality'] = mh_df['Dietary Quality'].map(dietmap)

mh_df
```

```python
mh_df.info()
```

```python
mh_df['Education Level'].value_counts()
```

```python
mh_df['Employment Status'].value_counts()
```

```python
employmap = {'Unemployed': 0, 'Employed': 1}
mh_df['Employment Status'] = mh_df['Employment Status'].map(employmap)

mh_df
```

```python
mh_subset = mh_df.sample(n=20000, random_state=84)

mh_subset
```

```python
mh_subset['Marital Status'].value_counts()
```

```python
mh_subset.rename(columns={'Marital Status': 'Ever Married'}, inplace=True)

mh_subset['Ever Married'] = mh_subset['Ever Married'].map(lambda x:1 if x in ['Married', 'Divorced', 'Widowed'] else 0)

mh_subset
```

```python
mh_subset['Smoking Status'].value_counts()
```

```python
smokemap = {'Current': 1, 'Non-smoker': 0, 'Former': 0}
mh_subset['Smoking Status'] = mh_subset['Smoking Status'].map(smokemap)

mh_subset
```

```python
mh_subset['Physical Activity Level'].value_counts()

#mh_subset['Physical Activity Level'] =
```

```python
actmap = {'Sedentary': 0, 'Moderate': 1, 'Active': 2}

mh_subset['Physical Activity Level'] = mh_subset['Physical Activity Level'].map(actmap)

mh_subset
```

```python
mh_subset.info()
```

```python
mh_subset.rename(columns={'History of Mental Illness': 'Mental Illness'}, inplace=True)

mh_subset
```

```python
mh_subset.to_csv('mental_health_prepped.csv', index=False)
```
